```python
import pandas as pd
import numpy as np
df = pd.read_csv("heart.csv")
df
```

```python
df.loc[:, 'zscore'] = (df.RestingBP - df.RestingBP.mean()) / df.RestingBP.std()
df = df[(df.zscore>-3)&(df.zscore<3)]
df.shape
```

```python
df.loc[:, 'zscore'] = (df.Cholesterol - df.Cholesterol.mean()) / df.Cholesterol.std()
df = df[(df.zscore>-3)&(df.zscore<3)]
df.shape
```

```python
df.loc[:, 'zscore'] = (df.Oldpeak- df.Oldpeak.mean()) / df.Oldpeak.std()
df = df[(df.zscore>-3)&(df.zscore<3)]
df.shape
```

```python
df.loc[:, 'zscore'] = (df.Age - df.Age.mean()) / df.Age.std()
df = df[(df.zscore>-3)&(df.zscore<3)]
df.shape
```

```python
df = df.drop(['zscore'] , axis = 'columns')
df.shape
```

```python
df = df.copy()
df.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df
```
