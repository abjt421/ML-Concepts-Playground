```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df1 = pd.DataFrame(df[['Pclass' , 'Sex' , 'Age' , 'Fare']])
df1
```

```python
dummies = pd.get_dummies(df['Sex']).astype(int)
dummies
```

```python
df2 = pd.concat([df1,dummies],axis='columns')
df2
```

```python
df2 = df2.drop(['Sex' , 'male'] , axis = 'columns')
df2
```

```python
df2.columns[df2.isna().any()]
```

```python
df2.Age[df2.Age.isna()]
```

```python
df2.Age = df2.Age.fillna(df2.Age.mean())
df2.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df2,df.Survived,test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```
