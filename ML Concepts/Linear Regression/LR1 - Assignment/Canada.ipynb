```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\abhij\Desktop\Skills\ML\ML Concepts\LR1 - Assignment\canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per capita income (US$)")
plt.scatter(df.year , df.pci , color = "black" , marker = "*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']] , df.pci)
```

```python
reg.predict([[2020]])
```
